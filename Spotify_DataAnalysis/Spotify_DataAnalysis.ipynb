```python
# Sort tracks based on popularity in ascending order
# finding 5 least popular songs
sorted_tracks = tracks_data.sort_values('popularity', ascending=True)
```

```python
# Another way to do it.
most_popular = tracks_data.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular.head()
```

```python
# Now lets make release date index of this dataset
tracks_data.set_index('release_date', inplace = True)
tracks_data.head()
```

```python
# Convert into datetime type
# This will convert Eg. 1922 to 1922-01-01
tracks_data.index = pd.to_datetime(tracks_data.index)
```

```python
tracks_data.head()
```

```python
# Convert miliseconds to seconds(timeduration)
tracks_data['duration'] = tracks_data['duration_ms'].apply(lambda x: round(x/1000))
tracks_data.drop(['duration_ms'], inplace=True, axis=1)
tracks_data.head()
```

```python
# This will drop and gives use pearson correclation dataframe
corr_df = tracks_data.drop(['key', 'mode', 'explicit'], axis = 1).corr(method='pearson')
corr_df
```

```python
# Now plot this data into a heatmap to get better visualization
plt.figure(figsize=(14,6))
heatmap = sb.heatmap(corr_df, vmin = -1, vmax = 1, annot=True)
heatmap.set_title("Correlation HeatMap between variables.")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

**Findings**: 
* Energy and loundness are very correlative. means if loundness is high then energy is also high with 76% Percent confidance.
* acousticness and energy are negatively correlative. means if acousticness is high then energy is also low with 72% Percent confidance.
* danceability and valence are positively correlative.
* And also some other relations we have.

```python
# Sampling .5% data to plot regression plot
sample_data = tracks_data.sample(frac =.005)
sample_data.shape
```

```python
# Now plot regression plot to see correlation between loudness and energy
plt.figure(figsize=(10,6))
sb.regplot(data= sample_data, y='loudness', x='energy', color='g')
plt.title("Loudness VS Energy Correlation")
plt.show()
```

**Findings:** 
* As we can see clearly if loudness is increasing the energy is also increasing and vice-versa.
* Also this is a linear realtionship.

```python
# Now plot regression plot to see correlation between acousticness and energy
plt.figure(figsize=(10,6))
sb.regplot(data= sample_data, y='acousticness', x='energy', color='c')
plt.title("Acousticness VS Energy Correlation")
plt.show()
```

**Findings:**
* As we can see this regression line is in the downtrend the it has negative relationship. Which is also represented in the heat map.

```python
tracks_data['dates'] = tracks_data.index.get_level_values('release_date')
tracks_data.dates = pd.to_datetime(tracks_data.dates)
years = tracks_data.dates.dt.year
```

```python
years # We got years
```

```python
# Now we are ploting distribution histogram for number of songs relaesed in each years
sb.displot(years, discrete = True, aspect =3, height=5)
plt.title('Number of songs released in each years')
plt.show()
```

```python
durations = tracks_data.duration # Getting durations
plt.figure(figsize=(18,6))
sb.barplot(x = years, y = durations)
plt.title('Duration VS Years')
plt.xticks(rotation=90)
plt.show()
```

**Findings:**
* As we can see duration of songs is varing and since 2010 it is in the downtrend.
